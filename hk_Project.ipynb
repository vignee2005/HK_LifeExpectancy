```python
# Analyzing and Visualizing Birth Rates Over Time
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df_birth = pd.read_csv('hkbirth.csv')
df_birth['Year'] = pd.to_datetime(df_birth['Year'])

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(df_birth['Year'], df_birth['Birth Rate (per 1000 people)'], marker='o', linestyle='-', color='b', label="Birth Rate")

# Add titles and labels
plt.title("Birth Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Birth Rate (per 1000 people)")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Analyzing and Visualizing Smoking Rates by Country and Gender
import matplotlib.pyplot as plt
import numpy as np

countries = ['India', 'China', 'United States']
rates_both = [27, 24.7, 25.1]
rates_male = [42, 47.7, 30.9]
rates_female = [12.1, 1.8, 19.3]

bar_width = 0.2
index = np.arange(len(countries))

fig, ax = plt.subplots()
bar1 = ax.bar(index - bar_width, rates_both, bar_width, label='Both')
bar2 = ax.bar(index, rates_male, bar_width, label='Male')
bar3 = ax.bar(index + bar_width, rates_female, bar_width, label='Female')

ax.set_xlabel('Countries')
ax.set_ylabel('Smoking Rates')
ax.set_title('Smoking Rates by Country and Gender')
ax.set_xticks(index)
ax.set_xticklabels(countries)
ax.legend()

plt.show()
```

```python
# Analyzing and Visualizing Overall Smoking Prevalence Over Time
import matplotlib.pyplot as plt

# Data
years = ['1982', '2002', '2021', '2025']
prevalence = [22.3, 14.4, 9.5, 7.8]

# Plotting the bar chart
plt.bar(years, prevalence, color='green')

# Adding title and labels
plt.title('Overall Smoking Prevalence')
plt.xlabel('Year')
plt.ylabel('Prevalence (%)')

# Display the arrow
for i, val in enumerate(prevalence[:-1]):
    plt.annotate('', xy=(i+1, prevalence[i+1]), xytext=(i, val),
                 arrowprops=dict(facecolor='blue', shrink=0.05),
                 )

# Highlighting the target
target_x = len(years) - 1
plt.scatter([target_x], [prevalence[target_x]], color='orange', zorder=5)
plt.text(target_x, prevalence[target_x]+1, 'Target', ha='center', color='blue')

# Show the plot
plt.show()
```

```python
# Analyzing and Visualizing Healthcare Costs Per Capita (2019-2022)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from your image, manually transcribed
data = {
    'country': ['China', 'United States', 'United Kingdom'],
    '2022': [None, 12555.3, 5492.6],
    '2021': [None, 12197, 5466.6],
    '2020': [978.8, 11915.9, 4997.9],
    '2019': [901.9, 10852.8, 4398]
}

# Convert the data into a Pandas DataFrame
df = pd.DataFrame(data)

# Set the country as the index for easier manipulation
df.set_index('country', inplace=True)

# Transpose the DataFrame to make it suitable for plotting
df = df.T

# Plot the data as a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.2
bar_positions = np.arange(len(df.index))

for i, country in enumerate(df.columns):
    ax.bar(bar_positions + i * bar_width, df[country], width=bar_width, label=country)

# Add data labels
for i, country in enumerate(df.columns):
    for j, year in enumerate(df.index):
        cost = df.at[year, country]
        if pd.notna(cost):
            ax.text(bar_positions[j] + i * bar_width, cost, f'{cost:.1f}', ha='center', va='bottom', fontsize=8, color='black')

# Add labels and title
ax.set_title('Healthcare Costs Per Capita (2019-2022)')
ax.set_xlabel('Year')
ax.set_ylabel('Healthcare Costs Per Capita ($)')
ax.set_xticks(bar_positions + bar_width * (len(df.columns) - 1) / 2)
ax.set_xticklabels(df.index)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
